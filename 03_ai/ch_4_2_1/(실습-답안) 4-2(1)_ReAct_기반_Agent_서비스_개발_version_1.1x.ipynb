### 핵심 개념
- **Agent**: 사용자 요청을 해석하고, 외부 도구를 활용하여 자율적으로 의사결정 및 행동을 수행하는 AI 시스템. LLM을 두뇌로 하여 환경과 상호작용한다.
- **Tool-learning**: LLM이 자연어 명령을 해석하여 적절한 외부 함수(API, DB 조회 등)를 선택하고 호출하는 능력. Function Calling이 대표적 구현 방식이다.
- **RAG (Retrieval-Augmented Generation)**: 외부 지식 베이스에서 관련 문서를 검색(Retrieve)하여 LLM의 컨텍스트에 추가함으로써 정확도와 신뢰성을 높이는 기법.
- **Memory**: Agent가 대화 이력(Short-term)과 사용자 정보(Long-term)를 저장하여 맥락을 유지하고 일관성 있게 응답하는 능력.
- **Plan**: Agent가 복잡한 작업을 하위 태스크로 분해하고, 실행 순서를 결정하며, 중간 결과를 반영하여 다음 행동을 조정하는 전략.
- **ReAct (Reasoning + Action)**: 추론(Thought)과 행동(Action)을 번갈아 수행하며, 관찰 결과(Observation)를 다음 추론에 반영하는 프레임워크.
- **LangGraph**: LangChain 생태계의 확장 라이브러리로, 상태 그래프(State Graph)를 사용하여 Agent의 워크플로우를 명시적으로 정의하고, 조건부 분기와 반복을 제어할 수 있다.
- **Trustworthiness**: Agent가 정책을 준수하고, 유해한 요청을 거부하며, 민감 정보를 보호하면서 안전하게 작동하도록 보장하는 설계 원칙.

### 3.1 정책 문서 준비

고객 서비스 Agent가 참조할 정책 문서를 준비합니다.

````python
# 정책 문서 생성 (실습용)
reward_policy = """
# 보상 정책 (Reward Policy)

## 쿠폰 발급 조건
1. 배송 지연 시: 최대 10,000원 쿠폰 발급 가능
2. 상품 불량 시: 최대 20,000원 쿠폰 발급 가능
3. 단순 변심: 쿠폰 발급 불가

## 쿠폰 발급 한도
- 1회 최대 발급 금액: 20,000원
- 월간 누적 한도: 50,000원

## 주의사항
- 배송 완료 후 7일 이내에만 보상 가능
- 이미 쿠폰을 받은 주문에는 추가 발급 불가
"""

shipping_policy = """
# 배송 정책 (Shipping Policy)

## 배송 상태
- 주문 접수: 주문이 접수된 상태
- 배송 준비중: 상품 포장 중
- 배송 중: 택배사에 인계됨
- 배송 완료: 고객에게 전달 완료
- 배송 지연: 예상 배송일을 초과함

## 배송 지연 기준
- 일반 배송: 결제 후 3영업일 초과 시 지연
- 새벽 배송: 주문 익일 미도착 시 지연
"""

# 파일로 저장
with open("reward_policy.txt", "w", encoding="utf-8") as f:
    f.write(reward_policy)

with open("shipping_policy.txt", "w", encoding="utf-8") as f:
    f.write(shipping_policy)

print("✅ 정책 문서 생성 완료")
````

### TODO 1: RAG 정책 검색 도구 구현

Vector Store를 구축하고 정책 검색 도구를 구현합니다.

- **요구사항**: retriever를 사용하여 정책 문서를 검색하고 결과를 문자열로 반환
- **입력**: `query` (검색할 질문 문자열)
- **출력**: 검색된 정책 내용 문자열
- **힌트**: `retriever.invoke(query)`로 문서를 검색하고, 결과를 `"\n\n".join()`으로 합치세요.

````python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma

# 문서 로드
loaders = [
    TextLoader("reward_policy.txt", encoding="utf-8"),
    TextLoader("shipping_policy.txt", encoding="utf-8")
]

documents = []
for loader in loaders:
    documents.extend(loader.load())

# 텍스트 분할
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
split_docs = text_splitter.split_documents(documents)

# Vector Store 생성
embeddings = OpenAIEmbeddings()

# Solar 대안 (주석 해제하여 사용)
# from langchain_upstage import UpstageEmbeddings
# embeddings = UpstageEmbeddings(model="embedding-query")

vectorstore = Chroma.from_documents(split_docs, embeddings)

# Retriever 생성
retriever = vectorstore.as_retriever(search_kwargs={"k": 2})

print(f"✅ Vector Store 생성 완료 (문서 {len(split_docs)}개)")
````

````python
# TODO 1: RAG 정책 검색 도구 구현
# 아래 함수를 완성하세요.

def search_policy(query: str) -> str:
    """
    정책 문서를 검색하여 관련 내용을 반환합니다.

    Args:
        query: 검색할 질문 (예: "쿠폰 발급 한도")

    Returns:
        검색된 정책 내용
    """
    # ===== 답안 =====
    # retriever를 사용하여 관련 문서 검색
    docs = retriever.invoke(query)

    # 검색 결과를 문자열로 합치기
    if docs:
        return "\n\n".join([doc.page_content for doc in docs])
    else:
        return "관련 정책을 찾을 수 없습니다."
    # ================
````

````python
# 정책 검색 테스트
result = search_policy("쿠폰 발급 한도")
print("검색 결과:")
print(result)
````

### 3.3 쿠폰 발급 도구 구현

실제 쿠폰을 발급하는 도구를 구현합니다.

````python
# 주문 데이터 (시뮬레이션용)
orders_db = {
    "ORD001": {"status": "배송 지연", "product": "노트북", "customer": "홍길동", "price" : 2000000},
    "ORD002": {"status": "배송 완료", "product": "키보드", "customer": "김철수", "price" : 200000},
    "ORD003": {"status": "배송 중", "product": "마우스", "customer": "이영희", "price" : 100000},
}

# 발급된 쿠폰 기록
issued_coupons = {}

def get_order_status(order_id: str) -> str:
    """주문 상태를 조회합니다."""
    if order_id in orders_db:
        order = orders_db[order_id]
        return f"주문번호: {order_id}, 상태: {order['status']}, 상품: {order['product']}, 가격 : {order['price']}"
    return f"주문 {order_id}을 찾을 수 없습니다."

def issue_coupon(order_id: str, amount: int) -> str:
    """
    쿠폰을 발급합니다. (검증 없는 기본 버전)
    쿠폰은 주문 상품의 5% 가격을 발급합니다.
    주의: 이 함수는 검증 로직이 없어 정책 위반이 발생할 수 있습니다.
    """
    issued_coupons[order_id] = amount
    return f"✅ 주문 {order_id}에 {amount:,}원 쿠폰 발급 완료"

print("✅ 도구 함수 정의 완료")
````

### 3.4 Tool 바인딩 및 테스트

LangChain의 Tool로 함수를 감싸고 LLM에 바인딩합니다.

````python
from langchain_core.tools import tool

# tool 데코레이터를 사용하여 도구 정의
@tool
def search_policy_tool(query: str) -> str:
    """고객 서비스 정책(보상, 배송)을 검색합니다. 쿠폰 발급 조건, 한도 등을 확인할 때 사용합니다."""
    return search_policy(query)

@tool
def get_order_status_tool(order_id: str) -> str:
    """주문 상태를 조회합니다. 주문번호(예: ORD001)를 입력하면 현재 상태를 반환합니다."""
    return get_order_status(order_id)

@tool
def issue_coupon_tool(order_id: str, amount: int) -> str:
    """쿠폰을 발급합니다. 주문번호와 금액을 입력합니다."""
    return issue_coupon(order_id, amount)

# Tool 목록 정의
tools = [search_policy_tool, get_order_status_tool, issue_coupon_tool]

print("✅ Tool 목록 정의 완료")
for tool in tools:
    print(f"  - {tool.name}: {tool.description[:50]}...")
````

### Tool 바인딩 메커니즘

```

Observation: 실행 결과 관찰
Thought: 관찰 결과를 바탕으로 다음 행동 결정
...반복...
Final Answer: 최종 응답
```

## Step 6: StateGraph 소개 - Graph 구현 도구

> **목표**: StateGraph의 핵심 개념을 이해하고, Agent 워크플로우를 그래프로 표현하는 방법 학습

### StateGraph란?

StateGraph는 LangGraph에서 제공하는 **Agent 워크플로우 정의 도구**입니다. 노드(Node)와 엣지(Edge)로 Agent의 실행 흐름을 명시적으로 정의할 수 있습니다.

```
┌─────────────────────────────────────────────────────────┐
│                    StateGraph 구조                       │
│                                                          │
│   [START] ──→ [Node A] ──→ [Node B] ──→ [END]          │
│                   │                                      │
│                   └──→ [Node C] ──→ [END]               │
│                                                          │
│   • 노드(Node): 실행할 함수                              │
│   • 엣지(Edge): 노드 간 연결 (무조건/조건부)             │
│   • 상태(State): 노드 간 공유되는 데이터                 │
└─────────────────────────────────────────────────────────┘
```

### StateGraph 핵심 컴포넌트

| 컴포넌트 | 설명 | 예시 |
|----------|------|------|
| **AgentState** | 그래프 전체에서 공유되는 상태 정의 | `messages`, `tool_calls` |
| **add_node()** | 노드(함수) 추가 | `workflow.add_node("agent", call_agent)` |
| **add_edge()** | 노드 간 연결 (무조건 이동) | `workflow.add_edge("tools", END)` |
| **add_conditional_edges()** | 조건부 분기 | 도구 호출 여부에 따라 분기 |
| **set_entry_point()** | 시작 노드 지정 | `workflow.set_entry_point("agent")` |
| **START, END** | 특수 노드 (시작/종료) | 그래프의 진입점과 종료점 |

### 왜 StateGraph를 직접 구현하는가?

Step 5에서 `create_react_agent`로 간편하게 Agent를 생성했습니다. 하지만 StateGraph를 직접 구현하면:

1. **내부 동작 이해**: `create_react_agent`가 어떻게 동작하는지 파악
2. **커스터마이징**: 노드/엣지 수준의 세밀한 제어 가능
3. **디버깅**: 각 노드에서 무슨 일이 일어나는지 추적 용이

````python
# =============================================================================
# StateGraph 공통 컴포넌트 정의
# - llm: Step 2에서 정의한 LLM 인스턴스
# - tools: Step 3에서 정의한 도구 목록
# =============================================================================

from langgraph.graph import StateGraph, START, END, MessagesState
from typing import TypedDict, Annotated, Sequence
from langchain_core.messages import BaseMessage, HumanMessage, AIMessage, ToolMessage
import operator



# # AgentState 정의: 그래프 전체에서 공유되는 상태
# MessagesState는 langgraph에서 가장 많이 쓰이는 상태 공유 방식
class AgentState(MessagesState):
    pass

# class AgentState(TypedDict):
#     """Agent의 상태를 정의합니다."""
#     messages: Annotated[Sequence[BaseMessage], operator.add]  # 메시지 히스토리 (누적)



print("✅ AgentState 정의 완료")
print("   - messages: 대화 히스토리를 저장하며, operator.add로 새 메시지가 누적됩니다.")
````

````python
# 노드 함수 정의: Graph에 등록할 함수들

def call_agent(state: AgentState) -> AgentState:
    """LLM을 호출하여 다음 행동을 결정합니다."""
    # LLM에 도구를 바인딩하여 호출
    # tool_choice="required" 는 tool 무조건 호출
    response = llm.bind_tools(tools).invoke(state["messages"])
    return {"messages": [response]}


def call_tools(state: AgentState) -> AgentState:
  # 마지막 메시지에서 llm 이 도구를 사용 할 것이라 판단 한다면 tool_calls에 사용할 도구를 설정함
  # llm이 직접 코드 실행을 하지 못하기 때문에 실행할 도구 이름만 적어서 놔

    """도구를 실행합니다."""
    last_message = state["messages"][-1]
    tool_messages = []

    # 각 도구 호출에 대해 실행
    for tool_call in last_message.tool_calls:
        for t in tools:
            if t.name == tool_call["name"]:
                result = t.invoke(tool_call["args"])
                tool_messages.append(ToolMessage(
                    content=str(result),
                    tool_call_id=tool_call["id"]
                ))
                break

    return {"messages": tool_messages}

print("✅ 노드 함수 정의 완료")
print("   - call_agent: LLM을 호출하여 응답 또는 도구 호출 결정")
print("   - call_tools: 선택된 도구를 실제로 실행")
````

### StateGraph 구성 준비 완료!

이제 다음 요소들이 준비되었습니다:
- **AgentState**: 그래프 전체에서 공유되는 상태
- **call_agent**: LLM 호출 노드 함수
- **call_tools**: 도구 실행 노드 함수

> **다음 Step에서**: 이 요소들을 조합하여 **Direct 패턴 Graph**를 구현합니다.

## Step 7: Direct Graph 구현

> **목표**: Step 6에서 준비한 컴포넌트로 **Direct 패턴 Graph**를 구현합니다.



### Concept Check: Direct Graph

**Direct Graph**는 도구를 1회만 실행하고 종료하는 단순한 워크플로우입니다. `tools → END` 엣지를 통해 반복 없이 즉시 결과를 반환합니다.

### Direct 패턴이란?

- **단일 패스(Single-pass)**: 도구를 1회 호출 후 바로 종료
- **빠른 응답**: 반복 없이 즉시 결과 반환
- **단순 작업에 최적화**: 주문 조회, 정책 검색 등 1-step 작업에 적합

### Direct 패턴 그래프 구조

```
[START] → [Agent] → [Tools] → [END]
              ↓
          (도구 없음) → [END]

※ 핵심: tools 후 바로 END (반복 없음)
```

### 구현 순서

1. **분기 함수 작성**: 도구 호출 여부 판단
2. **Graph 생성**: `StateGraph(AgentState)`
3. **노드 등록**: `add_node("agent", call_agent)`
4. **엣지 연결**: `add_edge()`, `add_conditional_edges()`
5. **컴파일**: `workflow.compile()`

5. 조건부 엣지: `add_conditional_edges()`
6. Direct 핵심: `add_edge("tools", END)` — 반복 없음
7. 컴파일: `workflow.compile()`

**힌트:**
- `should_continue_direct` 함수: 도구가 이미 실행되었거나 도구 호출이 없으면 `"end"` 반환

**예상 결과:**
- 그래프 구조: `[agent] → [tools] → [END]` (순환 없음)
- 테스트 시 도구 호출 횟수: **1회** (Direct 패턴의 특징)

**확인 방법**: `agent_direct.get_graph().draw_mermaid()` 실행 시 순환 없는 그래프 출력

````python
# TODO 2: Direct 패턴 StateGraph 구현

# ===== 답안 =====
# Direct 전용 분기 함수
def should_continue_direct(state: AgentState) -> str:
    """Direct: 도구 1회 실행 후 종료"""
    # 이미 도구가 실행되었으면 종료
    # for message in state["messages"]:
    #     if isinstance(message, ToolMessage):
    #        return "end"
    if any(isinstance(m, ToolMessage) for m in state["messages"]):
        return END

    # 도구 호출이 있으면 tools로, 없으면 종료
    last_message = state["messages"][-1]
    if hasattr(last_message, "tool_calls") and last_message.tool_calls:
      return "tools"
    return END
    # last = state["messages"][-1]
    # return "tools" if hasattr(last, "tool_calls") and last.tool_calls else "end"

# Direct 패턴 그래프 구성
workflow_direct = StateGraph(AgentState)

# 노드 추가
workflow_direct.add_node("agent", call_agent)
workflow_direct.add_node("tools", call_tools)

# 시작점 설정
# workflow_direct.set_entry_point("agent")
workflow_direct.add_edge(START, "agent")

# 조건부 엣지: agent → (tools 또는 end)
workflow_direct.add_conditional_edges(
    "agent",
    should_continue_direct,
    {"tools": "tools", END : END}
)

# ★ Direct 핵심: tools 후 바로 END (반복 없음)
workflow_direct.add_edge("tools", END)

# 그래프 컴파일
agent_direct = workflow_direct.compile()
# ================

print("✅ Direct 패턴 Agent 생성 완료")
````

````python
# Direct 패턴 그래프 시각화
print("=== Direct 패턴 그래프 구조 ===")
print(agent_direct.get_graph().draw_mermaid())

print("\n=== 그래프 구조 설명 ===")
print("""
[__start__] → [agent] → (조건 분기)
                           ├─ tool_calls 있음 → [tools] → [__end__]
                           └─ tool_calls 없음 → [__end__]

★ Direct 핵심: tools 후 바로 END (반복 없음)
""")
````

### Concept Check: ReAct Graph

**ReAct Graph**는 `tools → agent` 순환 엣지를 통해 추론과 행동을 반복합니다. 이를 통해 복잡한 멀티스텝 작업을 처리할 수 있습니다.

### ReAct 패턴이란?

ReAct(Reasoning + Action)는 **추론(Thought)**과 **행동(Action)**을 번갈아 수행하며, **관찰 결과(Observation)**를 다음 추론에 반영하는 프레임워크입니다.

### Direct vs ReAct 차이점

| 패턴 | 도구 실행 후 | 복잡한 요청 | 그래프 구조 |
|------|-------------|-------------|-------------|
| **Direct** | 바로 종료 | 처리 불가 | `tools → END` |
| **ReAct** | agent로 복귀 | 처리 가능 | `tools → agent` (순환) |

### ReAct 패턴 그래프 구조

```
        ┌──────────────┐
        │    START                   │
        └──────┬───────┘
                      ↓
        ┌──────────────┐
        │   Agent                    │ ←──────┐
        │  (LLM 호출)                │               │
        └──────┬───────┘               │
                      ↓                               │
        ┌──────────────┐               │
        │  조건 분기                 │               │
        ↓                            ↓               │
   ┌─────────┐  ┌─────────┐      │
   │  Tool            │  │   END            │      │
   │  실행            │  └─────────┘      │
   └────┬────┘                              │
             └────────────────────┘  ← ★ ReAct 핵심: 순환
```

### 구현 순서

1. **ReAct 분기 함수 작성**: Direct와 달리 "이미 실행됨" 체크 없음
2. **ReAct 시스템 프롬프트 작성**: Thought → Action → Observation 순환 유도
3. **Graph 구성**: `tools → agent` 순환 엣지 추가
4. **컴파일 및 테스트**

````python
# ReAct 시스템 프롬프트 (제공 코드)
# Thought → Action → Observation 순환을 유도하는 프롬프트

react_prompt_stategraph = """당신은 고객 서비스 AI 에이전트입니다.

다음의 ReAct 프레임워크를 따라 요청을 처리하세요:

1. Thought: 현재 상황을 분석하고 다음에 할 일을 생각합니다.
2. Action: 필요한 도구를 실행합니다.
3. Observation: 도구 실행 결과를 관찰합니다.
4. 위 과정을 반복하여 작업을 완료합니다.
5. Final Answer: 모든 작업이 완료되면 최종 응답을 제공합니다.

중요 원칙:
- 각 단계에서 Observation 결과를 반드시 확인하세요.
- 이전 결과를 바탕으로 다음 행동을 결정하세요.
- 정책 확인 → 주문 상태 확인 → 조건 판단 → 실행 순서를 지키세요."""

print("✅ ReAct 시스템 프롬프트 정의 완료")
````

### TODO 3: ReAct 분기 함수 + 순환 엣지 구현

ReAct 패턴의 핵심인 **분기 함수**와 **순환 엣지**를 구현하세요.

**ReAct의 핵심 (Direct와의 차이점):**
- **Direct**: `tools → END` (1회 실행 후 종료)
- **ReAct**: `tools → agent` (반복 가능 - 순환 구조)

**구현 요구사항:**

1. **분기 함수** (`should_continue_react`):
   - 도구 호출이 있으면 `"tools"` 반환
   - 도구 호출이 없으면 `"end"` 반환
   - Direct와 달리 "이미 실행됨" 체크 없음 → 계속 순환 가능

2. **ReAct 패턴 StateGraph 구성**:
   - 노드 추가: `add_node("agent", call_agent_react)`, `add_node("tools", call_tools)`
   - 시작점 설정: `set_entry_point("agent")`
   - 조건부 엣지: `add_conditional_edges()`
   - **★ ReAct 핵심**: `add_edge("tools", "agent")` — 순환 엣지

**힌트:**
- Direct의 `should_continue_direct`에서 "이미 ToolMessage가 있으면 종료" 로직을 제거하면 ReAct 분기 함수가 됩니다.
- Direct에서는 `add_edge("tools", END)`였지만, ReAct에서는 `add_edge("tools", "agent")`로 변경합니다.

**예상 결과:**
- 그래프 구조: `[agent] ↔ [tools]` (순환 구조)
- 복잡한 요청 테스트 시 도구 호출 횟수: **3회 이상**

**확인 방법**: `agent_react_stategraph.get_graph().draw_mermaid()` 실행 시 순환 구조의 그래프 출력

````python
# TODO 3: ReAct 분기 함수 + 순환 엣지 구현
from langchain_core.messages import SystemMessage

# call_agent_react 함수 (제공 코드)
def call_agent_react(state: AgentState) -> AgentState:
    """ReAct용 LLM 호출 (시스템 프롬프트 포함)"""
    messages = state["messages"]

    # 시스템 프롬프트가 없으면 추가
    if not any(isinstance(m, SystemMessage) for m in messages):
        messages = [SystemMessage(content=react_prompt_stategraph)] + list(messages)

    response = llm.bind_tools(tools).invoke(messages)
    return {"messages": [response]}

# ===== 답안 =====
# ReAct 전용 분기 함수 (Direct와 차이점: "이미 실행됨" 체크 없음)
def should_continue_react(state: AgentState) -> str:
    """ReAct: 도구 호출이 있으면 계속, 없으면 종료"""
    last_message = state["messages"][-1]

    # 도구 호출이 있으면 tools 노드로, 없으면 종료
    if hasattr(last_message, 'tool_calls') and last_message.tool_calls:
        return "tools"
    return END

# ReAct 패턴 그래프 구성
workflow_react = StateGraph(AgentState)

# 노드 추가
workflow_react.add_node("agent", call_agent_react)
workflow_react.add_node("tools", call_tools)

# 시작점 설정
# workflow_react.set_entry_point("agent")
workflow_react.add_edge(START,"agent")

# 조건부 엣지: agent → (tools 또는 end)
workflow_react.add_conditional_edges(
    "agent",
    should_continue_react,
    {"tools": "tools", END: END}
)

# ★ ReAct 핵심: tools 후 다시 agent로 (순환)
workflow_react.add_edge("tools", "agent")

# 그래프 컴파일
agent_react_stategraph = workflow_react.compile()
# ================

print("✅ ReAct 패턴 Agent 생성 완료")
````

### 해결: 검증 로직이 포함된 쿠폰 발급 함수

````python
# 검증 로직이 포함된 쿠폰 발급 함수 (제공 코드)

def extract_max_amount(policy_text: str) -> int:
    """LLM을 사용하여 정책 텍스트에서 1회 최대 발급 금액을 추출합니다."""
    prompt = f"""다음 정책 텍스트에서 '1회 최대 발급 금액'을 숫자만 추출하세요.
숫자만 반환하고, 다른 설명은 하지 마세요.

정책 텍스트:
{policy_text}

1회 최대 발급 금액 (숫자만):"""

    try:
        response = llm.invoke(prompt)
        # 응답에서 숫자만 추출
        amount_str = ''.join(filter(str.isdigit, response.content))
        return int(amount_str) if amount_str else 20000
    except Exception:
        return 20000  # 기본값 (추출 실패 시)

def issue_coupon_safe(order_id: str, amount: int) -> str:
    """
    신뢰성 있는 쿠폰 발급 (Trustworthiness 적용)

    검증 순서:
    1. 정책 조회 (RAG) → 발급 한도 확인
    2. 금액 검증 → 한도 초과 여부
    3. 주문 상태 확인 → 발급 조건 충족 여부
    4. 모든 검증 통과 → 실행
    """
    # 1. 정책 조회 (RAG 활용)
    policy_info = search_policy("쿠폰 발급 한도")

    # 정책에서 최대 금액 추출 (LLM 활용)
    max_amount = extract_max_amount(policy_info)

    # 2. 금액 검증
    if amount > max_amount:
        return f"❌ 발급 불가: 정책상 최대 {max_amount:,}원까지 가능합니다. (요청: {amount:,}원)"

    # 3. 주문 상태 확인
    if order_id not in orders_db:
        return f"❌ 발급 불가: 주문 {order_id}을 찾을 수 없습니다."

    status = orders_db[order_id]["status"]

    # 배송 지연 상태만 쿠폰 발급 가능
    if status != "배송 지연":
        return f"❌ 발급 불가: 현재 '{status}' 상태입니다. 배송 지연 시에만 쿠폰 발급이 가능합니다."

    # 4. 모든 검증 통과 → 실행
    issued_coupons[order_id] = amount
    return f"✅ 주문 {order_id}에 {amount:,}원 쿠폰 발급 완료 (정책 준수 확인됨)"

print("✅ 검증 로직이 포함된 쿠폰 발급 함수 정의 완료")
````

````python
# 검증 로직 테스트
print("=== 검증 로직 테스트 ===")

# 테스트 1: 정책 위반 금액
print("\n테스트 1 - 10만원 발급 시도 (한도 초과):")
print(issue_coupon_safe("ORD001", 100000))

# 테스트 2: 조건 미충족 (배송 완료)
print("\n테스트 2 - 배송 완료 주문에 발급 시도:")
print(issue_coupon_safe("ORD002", 5000))

# 테스트 3: 정상 발급
print("\n테스트 3 - 정상 발급 (배송 지연, 한도 내):")
print(issue_coupon_safe("ORD001", 5000))
````

### 가드레일: TrustedAgentExecutor 구현

````python
class TrustedAgentExecutor:
    """Trustworthiness가 적용된 Agent Executor"""

    def __init__(self, agent, allowed_tools=None):
        self.agent = agent
        self.allowed_tools = allowed_tools or []

        # 차단할 키워드 목록
        self.blocked_keywords = ["해킹", "비밀번호", "탈취", "시스템 접근"]

        # 민감 정보 패턴
        self.sensitive_patterns = ["주민등록번호", "계좌번호", "카드번호"]

    def is_harmful_input(self, user_input: str) -> bool:
        """입력이 유해한지 검사"""
        for keyword in self.blocked_keywords:
            if keyword in user_input:
                return True
        return False

    def filter_output(self, output: str) -> str:
        """출력에서 민감 정보 필터링"""
        filtered = output
        for pattern in self.sensitive_patterns:
            if pattern in filtered:
                filtered = filtered.replace(pattern, "[민감정보 필터링됨]")
        return filtered

    def run(self, user_input: str, config=None) -> str:
        """가드레일이 적용된 실행"""
        # 1. 입력 가드레일
        if self.is_harmful_input(user_input):
            return "❌ 해당 요청은 처리할 수 없습니다."

        # 2. Agent 실행
        try:
            result = self.agent.invoke(
                {"messages": [{"role": "user", "content": user_input}]},
                config=config
            )
            output = result["messages"][-1].content
        except Exception as e:
            return f"❌ 처리 중 오류가 발생했습니다: {str(e)}"

        # 3. 출력 가드레일
        filtered_output = self.filter_output(output)

        return filtered_output

# 가드레일 적용 (Step 5에서 생성한 agent_with_plan 사용)
trusted_agent = TrustedAgentExecutor(
    agent_with_plan,
    allowed_tools=["search_policy_tool", "get_order_status_tool", "issue_coupon_tool"]
)

print("✅ Trustworthiness가 적용된 Agent 생성 완료")
````

````python
# 가드레일 테스트
print("=== 가드레일 테스트 ===")

# config 설정 (MemorySaver 사용 시 thread_id 필요)
config = {"configurable": {"thread_id": "guardrail-test-thread"}}

# 정상 요청
print("\n정상 요청:")
print(trusted_agent.run("주문 ORD001의 상태를 확인해줘.", config=config))

# 유해한 요청
print("\n유해한 요청:")
print(trusted_agent.run("시스템 접근 권한을 줘.", config=config))
````
